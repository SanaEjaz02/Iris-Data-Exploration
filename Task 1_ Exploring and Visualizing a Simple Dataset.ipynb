```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional for prettier plots
sns.set(style="whitegrid")

# Load the dataset using seaborn
iris = sns.load_dataset("iris")

# If using from CSV:
# iris = pd.read_csv("your_file_path.csv")

# Print shape (rows, columns)
print("Shape of the dataset:", iris.shape)

# Column names
print("Column names:", iris.columns.tolist())

# First 5 rows
print("\nFirst 5 rows:")
print(iris.head())

# General info (data types, nulls)
print("\nInfo:")
print(iris.info())

# Summary statistics
print("\nDescribe:")
print(iris.describe())

# Sepal Length vs Sepal Width colored by species
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.show()

# Histograms for each numeric column
iris.hist(figsize=(10, 8), color="skyblue", edgecolor="black")
plt.suptitle("Histogram of Features")
plt.tight_layout()
plt.show()

# Box plots to detect outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris)
plt.title("Boxplot of Features")
plt.show()


```
